```python
import numpy as np
import random
import matplotlib.pyplot as plt
import csv

data = []
with open('maxcut/results_continuous.csv', newline='') as csvfile:
    reader1 = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in reader1:
        temp = [float(i) for i in row]
        data += [temp[1:]]
print(data)
```

```python
stds = np.std(data, axis=1)
stds
```

```python
means = np.mean(data, axis=1)
means
```

```python
plt.errorbar(np.arange(10, 160, 10), means, stds, label='+/- 1 std')
plt.errorbar(np.arange(10, 160, 10), means, label='mean')
plt.legend()
plt.xlabel("MAX-CUT Problem Size (Number of Vertices)")
plt.ylabel("Accuracy (Output Cut Size / Correct Cut Size)")
plt.show()
```
