```python
import numpy as np
import random
import matplotlib.pyplot as plt
import csv
```

```python
# csv_names = ['continuous.csv', 'discrete.csv', 'continuous_rbm.csv', 'discrete_rbm.csv']
csv_names = ['continuous_rbm.csv', 'discrete_rbm.csv']
for csv_name in csv_names:
    data = []
    with open('maxcut/' + csv_name, newline='') as csvfile:
        reader1 = csv.reader(csvfile, delimiter=' ', quotechar='|')
        for row in reader1:
            temp = [float(i) for i in row]
            data += [temp[1:]]
    stds = np.std(data, axis=1)
    means = np.mean(data, axis=1)
    plt.errorbar(np.arange(10, 160, 10), means, stds, label='+/- 1 std')
    plt.errorbar(np.arange(10, 160, 10), means, label='mean')
    plt.legend()
    plt.xlabel("MAX-CUT Problem Size (Number of Vertices) for " + csv_name)
    plt.ylabel("Accuracy (Output Cut Size / Correct Cut Size)")
    plt.show()
```
